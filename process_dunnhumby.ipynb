# Process `dunnhumby`

We downsize the `dunnhumby` dataset to only those associated with store ids `367`, `406`, `356` in order to fit Google Sheets.

```python
import pandas as pd
```

## `transaction_data`

```python
transaction_data = pd.read_csv('journey/transaction_data.csv')
transaction_data
```

```python
transaction_data.groupby(['STORE_ID']).SALES_VALUE.agg([sum,len]).reset_index().sort_values('len',ascending=False)
```

```python
store_ids = [367,406,356]
transaction_data = transaction_data[transaction_data.STORE_ID.isin(store_ids)].reset_index(drop=True)
transaction_data.to_csv('journey_mini/transaction_data.csv',index=False)
transaction_data.shape
```

## `hh_demographic`

```python
hh_demographic = pd.read_csv('journey/hh_demographic.csv')
print(hh_demographic.shape)
hh_demographic = hh_demographic[hh_demographic.household_key.isin(list(transaction_data.household_key.unique()))]
hh_demographic.to_csv('journey_mini/hh_demographic.csv',index=False)
hh_demographic.shape
```
